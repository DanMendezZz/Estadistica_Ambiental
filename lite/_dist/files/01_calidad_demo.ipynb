```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(0)
n = 720
fechas = pd.date_range("2024-01-01", periods=n, freq="h")
ciclo = 25 + 10 * np.sin(2 * np.pi * np.arange(n) / 24)
ruido = rng.normal(0, 4, size=n)
pm25 = np.clip(ciclo + ruido, 0, None)
# Inyectamos faltantes y un outlier para que assess_quality tenga algo que reportar.
pm25[100:120] = np.nan
pm25[300] = 250.0
df = pd.DataFrame({"fecha": fechas, "estacion": "demo", "pm25": pm25})
df.head()
```

## `assess_quality` — diagnóstico EDA estandarizado

```python
from estadistica_ambiental.eda.quality import assess_quality

calidad = assess_quality(df, date_col="fecha")
print(calidad.summary())
```

## `summarize` — resumen descriptivo univariado

```python
from estadistica_ambiental.descriptive.univariate import summarize

summarize(df[["pm25"]]).round(2)
```
