# 02 — Tendencia con Mann-Kendall y Sen

Demo en navegador del test no-paramétrico de tendencia sobre una serie ambiental sintética.

Mann-Kendall responde: ¿hay tendencia monótona estadísticamente significativa? Sen estima la
magnitud (pendiente robusta a outliers).

```python
import piplite
await piplite.install(["estadistica-ambiental", "pymannkendall"])
```

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(42)
n = 240  # 20 años de datos mensuales
fechas = pd.date_range("2005-01-01", periods=n, freq="MS")
tendencia = 0.05 * np.arange(n)         # +0.05 unidades por mes
estacional = 3 * np.sin(2 * np.pi * np.arange(n) / 12)
ruido = rng.normal(0, 1.0, size=n)
valor = 20 + tendencia + estacional + ruido
serie = pd.Series(valor, index=fechas, name="caudal")
serie.head()
```

## `mann_kendall` — test de tendencia + Sen slope

```python
from estadistica_ambiental.inference.trend import mann_kendall

resultado = mann_kendall(serie)
resultado
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(9, 3.5))
serie.plot(ax=ax, label="serie observada")
ax.set_title(f"Tendencia Mann-Kendall — Sen slope ≈ {resultado['slope']:.3f}/mes ({resultado['trend']})")
ax.set_ylabel("caudal (unidades sintéticas)")
ax.legend()
plt.show()
```
